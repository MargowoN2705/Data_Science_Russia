```python
# Funkcja ujednolicająca format nazw regionów
def standardize_region(region):
    region = region.strip().lower()  # Ujednolicenie wielkości liter i usunięcie zbędnych spacji
    region = region.replace("[a]", "")  # Usunięcie zbędnych znaków (np. '[a]')
    return region

# Ujednolicenie regionów w `pkb`
pkb['Region'] = pkb['Region'].apply(standardize_region)

# Ujednolicenie regionów w `amount`
amount['Region'] = amount['Region'].apply(standardize_region)

# Sprawdzamy, czy zbiory są teraz identyczne
set3 = set(amount['Region'].unique())
set4 = set(pkb['Region'].unique())

if set3 == set4:
    print("Zbiory są identyczne.")
else:
    print("Zbiory się różnią.")
    diff3 = set3 - set4
    print("Regiony w amount, których nie ma w pkb:\n", diff3)
    print(f'{len(diff3)} \n')

    diff4 = set4 - set3
    print("Regiony w pkb, których nie ma w amount:\n", diff4)
    print(f'{len(diff4)} \n')
```

```python
# Usuwanie pustych wartości
pkb = pkb[pkb['Region'].str.strip() != '']

# Ujednolicanie dalszych różnic w regionach
def clean_region_names(region):
    region = region.strip().lower()
    region = region.replace("st. petersburg", "saint petersburg")  # Ujednolicenie
    region = region.replace("north ossetia", "north ossetia–alania")  # Dostosowanie regionu
    return region

# Zastosowanie czyszczenia do danych
pkb['Region'] = pkb['Region'].apply(clean_region_names)
set3 = set(amount['Region'].unique())
set4 = set(pkb['Region'].unique())

if set3 == set4:
    print("Zbiory są identyczne.")
else:
    print("Zbiory się różnią.")
    diff3 = set3 - set4
    print("Regiony w amount, których nie ma w pkb:\n", diff3)
    print(f'{len(diff3)} \n')

    diff4 = set4 - set3
    print("Regiony w pkb, których nie ma w amount:\n", diff4)
    print(f'{len(diff4)} \n')

```

```python
import re

# Funkcja do czyszczenia nazw regionów
def clean_region_name(region_name):
    # Usuwamy słowa typu 'oblast', 'krai', 'autonomous okrug', itp.
    region_name = re.sub(r'( oblast| krai| autonomous okrug| republic| federal district| oblast| a.o.| city)', '', region_name, flags=re.IGNORECASE)
    # Zmieniamy na małe litery, aby porównanie było bezwzględne względem wielkości liter
    region_name = region_name.lower().strip()
    return region_name

# Ujednolicenie nazw regionów w obu zbiorach
set3_cleaned = set(clean_region_name(region) for region in amount['Region'].unique())
set4_cleaned = set(clean_region_name(region) for region in pkb['Region'].unique())

# Porównanie ujednoliconych zbiorów
if set3_cleaned == set4_cleaned:
    print("Zbiory są identyczne.")
else:
    print("Zbiory się różnią.")
    diff3 = set3_cleaned - set4_cleaned
    print("Regiony w amount, których nie ma w pkb:\n", diff3)
    print(f'{len(diff3)} \n')

    diff4 = set4_cleaned - set3_cleaned
    print("Regiony w pkb, których nie ma w amount:\n", diff4)
    print(f'{len(diff4)} \n')
```

```python
import re

# Funkcja do czyszczenia nazw regionów
def clean_region_name(region_name):
    # Usuwamy słowa typu 'oblast', 'krai', 'autonomous okrug', itp.
    region_name = re.sub(r'( oblast| krai| autonomous okrug| republic| federal district| oblast| a.o.| city)', '', region_name, flags=re.IGNORECASE)

    # Usuwamy nadmiarowe powtórzenia tego samego słowa
    region_name = ' '.join(sorted(set(region_name.split()), key=region_name.split().index))

    # Zmieniamy na małe litery, aby porównanie było bezwzględne względem wielkości liter
    region_name = region_name.lower().strip()

    return region_name

# Ujednolicenie nazw regionów w obu zbiorach
set3_cleaned = set(clean_region_name(region) for region in amount['Region'].unique())
set4_cleaned = set(clean_region_name(region) for region in pkb['Region'].unique())

# Porównanie ujednoliconych zbiorów
if set3_cleaned == set4_cleaned:
    print("Zbiory są identyczne.")
else:
    print("Zbiory się różnią.")
    diff3 = set3_cleaned - set4_cleaned
    print("Regiony w amount, których nie ma w pkb:\n", diff3)
    print(f'{len(diff3)} \n')

    diff4 = set4_cleaned - set3_cleaned
    print("Regiony w pkb, których nie ma w amount:\n", diff4)
    print(f'{len(diff4)} \n')
```
