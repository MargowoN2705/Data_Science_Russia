```python
data = pd.read_csv('/content/merged_data.csv')
data
```

```python
data = data.drop('Polulation', axis=1)
data
```

```python
data = data.rename(columns={'PKB': 'GRDP_per_capita'})

```

```python
data['GRDP_per_capita'] = pd.to_numeric(data['GRDP_per_capita'], errors='coerce')
```

```python

regions_to_divide = ['moscow', 'arkhangelsk oblast\n(except ao)', 'krasnoyarsk krai', 'kamchatka krai', 'irkutsk oblast']

data.loc[data['Region'].isin(regions_to_divide), 'GRDP_per_capita'] = data.loc[data['Region'].isin(regions_to_divide), 'GRDP_per_capita'] / 1000
```
